```python
#Martín Samuel Calderón Ojeda
#IDS TM 9No
#Red neuronal en python para resolver el problema de XOR

#Librerias para realizar operaciones matematicas
import random
import math


#funciones de activacion
def sigmoide(x):
    return 1.0 / (1.0 + math.exp(-x))


def sigmoide_prime(x): # x already sigmoided
    return x * (1 - x)

def aprendizaje(entradas,targets,alpha=0.1):
    global pesos_ocultas, sesgo_ocultas
    global pesos_salidas, sesgo_salidas
    
    hiddens = []
    for i in range(OCULTA):
        hidden = 0
        for j in range(ENTRADA):
            hidden += pesos_ocultas[i][j] * entradas[j]
        hidden = sigmoide(hidden + sesgo_ocultas[i])
        hiddens.append(hidden)
    
    salidas = []
    for i in range(SALIDA):
        salida = 0
        for j in range(OCULTA):
            salida += pesos_salidas[i][j] * hiddens[j]
        salida = sigmoide(salida + sesgo_salidas[i])
        salidas.append(salida)
    
    errores = []
    for i in range(SALIDA):
        error = targets[i] - salidas[i]
        errores.append(error)
    
    derrors = []
    for i in range(SALIDA):
        derror = errores[i] * sigmoide_prime(salidas[i])
        derrors.append(derror)
    
    ds = [0] * OCULTA
    for i in range(SALIDA):
        for j in range(OCULTA):
            ds[j] += derrors[i] * pesos_salidas[i][j] * sigmoide_prime(hiddens[j])
    
    for i in range(SALIDA):
        for j in range(OCULTA):
            pesos_salidas[i][j] += alpha * hiddens[j] * derrors[i]
        sesgo_salidas[i] += alpha * derrors[i]
    
    for i in range(OCULTA):
        for j in range(ENTRADA):
            pesos_ocultas[i][j] += alpha * entradas[j] * ds[i]
        sesgo_ocultas[i] += alpha * ds[i]


def prediccion(entradas):
    hiddens = []
    for i in range(OCULTA):
        hidden = 0
        for j in range(ENTRADA):
            hidden += pesos_ocultas[i][j] * entradas[j]
        hidden = sigmoide(hidden + sesgo_ocultas[i])
        hiddens.append(hidden)
    
    salidas = []
    for i in range(SALIDA):
        salida = 0
        for j in range(OCULTA):
            salida += pesos_salidas[i][j] * hiddens[j]
        salida = sigmoide(salida + sesgo_salidas[i])
        salidas.append(salida)
    
    return salida

#Definir datos red
VARIACION_PESO = 0.5

ENTRADA = 2
OCULTA = 3
SALIDA = 1

pesos_ocultas = []
for _ in range(OCULTA):
    pesos_ocultas.append([random.uniform(-VARIACION_PESO,VARIACION_PESO) for _ in range(ENTRADA)])

sesgo_ocultas = [0] * OCULTA

pesos_salidas = []
for _ in range(SALIDA):
    pesos_salidas.append([random.uniform(-VARIACION_PESO,VARIACION_PESO) for _ in range(OCULTA)])
    
sesgo_salidas = [0] * SALIDA


#Entrenamiento entradas
entradas = [[0,0],[0,1],[1,0],[1,1]]

#Entrenamiento salidas
salidas = [[0],[1],[1],[0]]


for i in range(10000):
    indexes = [0,1,2,3]
    random.shuffle(indexes)
    for j in indexes:
        aprendizaje(entradas[j],salidas[j],alpha=0.2)
    
    if (i+1) % 1000 == 0:
        cost = 0
        for j in range(4):
            o = prediccion(entradas[j])
            cost += (salidas[j][0] - o) ** 2
        cost /= 4
        print("Iteracion: ",i+1)      
        print("Error:", cost) 





#Impresiones finales
for i in range(4):
    resultado = prediccion(entradas[i])
    print("Entrada", entradas[i], "Salida Esperada: ", salidas[i][0], "| Prediccion:", f"{resultado:4.4}", ": Por tanto es", "Correcto" if round(resultado)==salidas[i][0] else "incorrecto")
```
